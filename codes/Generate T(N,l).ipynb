leaf $(A,n)$ function in the following cell chooses randomly a leaf node among the nodes of input tree graph with adjacency matrix A and size n:

non_leaf(A,n) function in the following cell chooses randomly a non_leaf node among the nodes of input tree graph with adjacency matrix A and size n:

In both functions, outputs are a number which determines the label of the selected node in the adjacency matrix.

```python
import numpy as np
import random
import math

def leaf(A,n):
    L=0
    while L==0:
        t=0
        u=random.randint(0,n-1)
        for j in range (0,n):
            t=t+A[u,j]
        if t==1:
            break
    return(u)


def non_leaf(A,n):
    L=0
    while L==0:
        t=0
        u=random.randint(0,n-1)
        for j in range (0,n):
            t=t+A[u,j]
        if t>1:
            break
    return(u)
```

## Generating 
In the folowing cell using leaf(A,n) and non_leaf(A,n) functions, function TR(n,l) constructs a tree with n nodes which l of them are leaves. The output is the adjacency matrix of the graph. Since there are many tree graphs with sizes n and l leaves, therefore, the output is not unique.




```python
def Tre(n,l):
    s=(n,n)
    A=np.zeros(s)    
    A[0,1]=1
    A[1,0]=1
    A[1,2]=1
    A[2,1]=1
    if n-l-1>l-2:
        K=(n-l-1)//(l-2)
        S=(K+1)*(l-2)+1
        for i in range (3,S):
            if i%(K+1)==0:
                u=non_leaf(A,n)
                A[u,i]=1
                A[i,u]=1
            else:
                g=leaf(A,n)
                A[g,i]=1
                A[i,g]=1        
        for j in range (S,n):
            l=leaf(A,n)
            A[l,j]=1
            A[j,l]=1
        return(A)
    else:
        K=(l-2)//(n-l-1)
        S=(K+1)*(n-l-1)+1
        for i in range (3,S):
            if i%(K+1)==0:
                u=leaf(A,n)
                A[u,i]=1
                A[i,u]=1
            else:
                g=non_leaf(A,n)
                A[g,i]=1
                A[i,g]=1        
        for j in range (S,n):
            l=non_leaf(A,n)
            A[l,j]=1
            A[j,l]=1
        return(A)
```

Function  Moran_FT(n,l,r,Q) runs the Moran process upon a tree graph generated by function T(N,l). Since T(N,l) is not unique Moran_FT(n,l,r,Q) average over Q realization.   r determines the fitness of simulation. In the second line of the below cell, one can substitute Tre(n,l) by adjacency matrix of any graph and run the Moran process upon that graph.

Function Mean_FT(n,l,r,q) runs the   Moran_FT(n,l,r)  for q times and averages over q output.  

```python
def Moran_FT(n,l,r,Q):
    A=Tre(n,l)
    T=0    
    for x in range (Q):
        B=[1]
        for k in range (0,n-1):
            B.append(1)
        l=random.randint(0,n-1)
        B[l]=-1
        t=0
        i=1
        while i<n:
            k=random.random()
            if k<(i*r)/(n-i+i*r):
                s=random.randint(0,n-1)
                j=random.randint(0,n-1)
                while B[s]!=-1:
                    s=random.randint(0,n-1)
                while A[s,j]!=1:
                    j=random.randint(0,n-1)    
                if B[j]==1:
                    t=t+1
                    i=i+1 
                    B[j]=-1
                else:
                    t=t+1
            else:
                w=random.randint(0,n-1)
                q=random.randint(0,n-1)
                while B[w]!=1:
                    w=random.randint(0,n-1)
                while A[w,q]!=1:
                    q=random.randint(0,n-1)
                if B[q]==-1:
                    if i==1:
                        t=0
                    else:
                        t=t+1
                        i=i-1 
                        B[q]=1
                else:
                    t=t+1 
        T=T+t
    return(T/Q)

def Mean_FT(n,l,r,q):
    it=10
    for j in range (0,q):
        tt=tt+Moran_FT(n,l,r) 
    return(tt/q)
         
```

 function one_tree(n) return the  adjacency matrix matrix of  1-tree graph with size n.

```python
def one_tree(n):
    s=(n,n)
    A=np.zeros(s)    
    A[0,1]=1
    A[1,0]=1
    A[1,2]=1
    A[2,1]=1
    for i in range (3,n):
        u=random.randint(0,i-1)
        A[i,u]=1
        A[u,i]=1
    return(A)
```
